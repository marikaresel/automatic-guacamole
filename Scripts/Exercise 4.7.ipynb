# Task 4.7

```python
result = []

# Determine the two busiest and two slowest days
busiest_days = ords_prods_merge["orders_day_of_week"].value_counts().nlargest(2).index.tolist()
slowest_days = ords_prods_merge["orders_day_of_week"].value_counts().nsmallest(2).index.tolist()

for value in ords_prods_merge["orders_day_of_week"]:
    if value in busiest_days:
        result.append("Busiest days")
    elif value in slowest_days:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")

# Create a new column with the updated labels
ords_prods_merge["busiest_days"] = result
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# Determine the busiest, average, and least busy hours of the day
hourly_order_counts = ords_prods_merge['order_hour_of_day'].value_counts()

most_orders_hours = hourly_order_counts[hourly_order_counts == hourly_order_counts.max()].index.tolist()
fewest_orders_hours = hourly_order_counts[hourly_order_counts == hourly_order_counts.min()].index.tolist()

# For average orders, we can consider the hours between most and least busy hours
average_orders_hours = [hour for hour in range(24) if hour not in most_orders_hours and hour not in fewest_orders_hours]

# Create a new column with appropriate labels
ords_prods_merge['busiest_period_of_day'] = ords_prods_merge['order_hour_of_day'].apply(
    lambda hour: 'Most orders' if hour in most_orders_hours else (
        'Average orders' if hour in average_orders_hours else 'Fewest orders')
)
```

```python
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
