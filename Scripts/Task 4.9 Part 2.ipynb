```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Import data file "orders_products_customers_combined.pkl"
ords_prods_customers_merge = pd.read_pickle('/Users/marika/Instacart Basket Analysis 04.14.24/02 Data/Prepared Data/orders_products_customers_merged.pkl')
```

```python
# Import data file "customers.csv"
df_customers = pd.read_csv('/Users/marika/Instacart Basket Analysis 04.14.24/02 Data/Original Data/customers.csv')
```

```python
# Plot a histogram of the "order_hour_of_day" column
plt.figure(figsize=(10, 6))
plt.hist(ords_prods_customers_merge['order_hour_of_day'], bins=24, color='skyblue', edgecolor='black')
plt.title('Distribution of Order Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Frequency')
plt.xticks(range(24))  # Set x-axis ticks to hours of the day
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# the histogram demonstrates clear patterns in order timing, reflecting typical consumer behavior and preferences throughout the day.
```

```python
# Plot a bar chart of the distribution of orders based on loyalty_flag
plt.figure(figsize=(8, 6))
ords_prods_customers_merge['loyalty_flag'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Orders by Loyalty Flag')
plt.xlabel('Loyalty Flag')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group by hour of the day and calculate the average expenditure
hourly_expenditure = ords_prods_customers_merge.groupby('order_hour_of_day')['prices'].mean()

# Plot the average expenditure by hour of the day
plt.figure(figsize=(10, 6))
hourly_expenditure.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Expenditure by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Expenditure ($)')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group the data by age and calculate the average number of dependents
avg_dependents_by_age = df_customers.groupby('Age')['n_dependants'].mean()

# Plot a line chart of the average number of dependents by age
plt.figure(figsize=(10, 6))
avg_dependents_by_age.plot(kind='line', color='skyblue', marker='o')
plt.title('Average Number of Dependents by Age')
plt.xlabel('Age')
plt.ylabel('Average Number of Dependents')
plt.grid(True)
plt.show()
```

```python
# The line chart shows some fluctuations in the average number of dependents across different age groups, 
indicating variations in family sizes and structures among customers.
```

```python
# Create a scatterplot of age vs. income
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_customers, x='Age', y='income', color='skyblue')
plt.title('Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```

```python
# The scatterplot highlights the diversity of income levels among customers of different ages, suggesting that age alone may not 
be a reliable indicator of spending power.
```
